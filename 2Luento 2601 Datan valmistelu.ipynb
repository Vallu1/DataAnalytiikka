<h2>Datan avaaminen</h2>

Excel-muotoisen datan voin lukea pandas-kirjaston <b>read_excel()</b>-funktiolla. <b>head()</b>-funktio tulostaa datan viisi ensimmäistä riviä, <b>tail()</b>-funktio tulostaa viisi viimeistä riviä ja <b>shape</b>-ominaisuus antaa datan rivien ja sarakkeiden lukumäärän.

```python
# Avattu data sijoitetaan df muuttujan arvoksi. 
# Jos data ei ole ensimmäisellä välilehdellä tarvitaan lisäparametri sheet_name ja välilehden nimi.
df=pd.read_excel('http://taanila.fi/data1.xlsx', sheet_name='Data')
df.head()
```

Phyton luo automaattisesti vasempaan reunaan index sarakkeen (joka ei ole ladattavassa datassa mitenkään mukana). Indexin avulla voidaan paikallistaa ja yksilöidä tietoja. 
Jos sarakkeessa on tyhjiä soluja, python tulkitsee sarakkeen desimaaliluvuksi vaikka siellä olisi kaikki luvut kokonaislukuja. Jos kaikki sarakkeen luvut ovat kokonaislukuja eikä siellä joukossa ole yhtään tyhjää solua - python tulkitsee ne kokonaisluvuiksi.

```python
df.tail(10)
```

```python
#Dataframen rivien ja sarkkeiden määrä, saadaan selville shape toiminnolla 82 riviä ja 16 saraketta
df.shape
```

<h2>Sarakkeet</h2>

Sarakkeiden (muuttujien) nimilistan saan tulostamalla <b>columns</b>-ominaisuuden. Voin vaihtaa muuttujien nimiä sijoittamalla muutetun nimilistan columns-ominaisuuden arvoksi. Muuttujien nimiä voin vahtaa myös <b>rename()</b>-funktiolla.

```python
df.columns
```

```python
# Lista on pythonissa aina hakasulkeiden välissä [ ..., ...] 
df.columns=['nro', 'sukupuoli', 'ikä', 'perhe', 'koulutus', 'palveluv', 'palkka',
       'johto', 'työtov', 'työymp', 'palkkat', 'työteht', 'työterv', 'lomaosa',
       'kuntosa', 'hieroja']
df.columns
```

```python
#Vaihdan muuttujien nimiä rename komennolla. inplace=True komennolla tallennetaan muutokset kantaan
# sanakirja (dictionary) on aina pythonissa aaltosulkeiden { ..., ... } välissä 
# kaarisulkeiden välissä on eri komentojen ja funktioiden parametrejä. parametrit erotetaan toisistaan pilkulla  
df.rename(columns={'sukupuoli':'sukup','työtov':'toveri'},inplace=True)
df.columns
```

Muuttujien arvojen lukumäärät selviävät <b>count()</b>-funktiolla.
- se näyttää sarakkeiden nimet ja tyhjästä poikkeavien arvojen lukumäärän

```python
df.count()
```

<h2>Ainutkertaiset arvot</h2>

Muuttujan sisältämät ainutkertaiset arvot selviävät <b>unique()</b>-funktiolla. Seuraavassa käydään läpi kaikki muuttujat <b>for</b>-silmukalla.

Tässä for ... in -silmukassa käydään läpi kaikki muuttujat df aineistosta.
- var:in tilalla voisi olla mikä tahansa muuttujan nimi 
- pythonissa " : " aloittaa silmukan ja kaikki mikä on sisennyksessä kuuluvat for silmukkaan 
- kun sisennys loppuu loppuu myös for silmukka

```python
# Tässä for ... in -silmukassa käydään läpi kaikki muuttujat df aineistosta.
# jokaisen muuttujan kohdalla printataan näytölle yksittäiset muuttujan arvot 
# var :ssa on muuttujan nimi (=sarake otsikko)
# toisena tietona tulostetaan kaikki yksittäiset muuttujan arvot pd.unique funktiolla
# unique on funktio jollon sille tarvitaan antaa jotain parametrinä 
# ja sille annetaan parametrinä sieltä datasta se muuttuja 
# tässä on hakasulkeille toinen käyttötarkoitus, aina kun haetaan datan sisältä jotain 
# tiettyä muuttujaa niin sen muuttujan nimi laitetaan aina hakasulkujen sisään. 

for var in df:
    print(var, pd.unique(df[var]))
```

<h2>Uusien muuttujien laskeminen</h2>

<b>replace()</b>-funktiolla korvaan muuttujan arvoja uusilla arvoilla.
<b>cut()</b>-funktiolla voin luokittelen määrällisen muuttujan arvot luokkiin.
<b>mean()</b> ja <b>sum()</b> -funktioilla voin laskea keskiarvo- ja summamuuttujia (<b>axis=1</b> tarkoittaa laskemista rivien suuntaisesti).

```python
# Uusi muuttuja 'sukup_teksti', jossa sukupuolet tekstiarvoina. Aina on fiksua säästää org.tieto=sukup
# HUOM ! Kokonaislukuja ei laiteta heittomerkkien sisään 
df['sukup_teksti']=df['sukup'].replace({1:'Mies', 2:'Nainen'})
df.head(6)
```

```python
# Uusi muuttuja 'ikäluokka', jossa iät luokiteltuina ikäluokkiin
# binssit määrittää luokkarajat ja ne määritellään listana 
# (18, 28] = tarkoittaa että 18v ei kuulu luokkaan ja 28] -tarkoittaa että 28v kuuluu luokkaan
# kaarisulku ( tarkoittaa että ei kuulu luokkaan ja hakasulku [ tarkoittaa että kuuluu
# muuten on vaikea tehdä frekvenssitaulukkoa
bins=[18, 28, 38, 48, 58, 68]
df['ikäluokka'] = pd.cut(df['ikä'], bins = bins)
df.head()
```

```python
# Lasketaan kokonaistyytyväisyys. 
# Tehdään uusi muuttuja = tyytyväisyys ja lasketaan keskiarvona muuttujista johto, toveri, jne..
# Lista pistetään omiin hakasulkuihin 
# Nyt kun lasketaan riveiltä keskiarvoa sen takia lisämääritys -> axis = 1
df['tyytyväisyys'] = df[['johto','toveri','työymp','palkkat','työteht']].mean(axis=1)
df.head(6)
```

```python
# Kuinka montaa asiaa kukin käyttää? - tehdään uusi muuttuja 'käyttö' ja nyt summataan sum funktio. 
df['käyttö']=df[['työterv','lomaosa','kuntosa','hieroja']].sum(axis=1)
df.head()
```

<h2>Poimintaa ja suodatuksia</h2>

Voin muodostaa monimutkaisiakin poimintoja <b>&</b>- ja <b>|</b>-operaattoreita käyttäen.

<b>notnull()</b>- ja <b>isnull()</b>-funktioilla voin selvittää ei-tyhjät ja tyhjät.

<b>sort_values()</b>-funktio järjestää halutun muuttujan suhteen.

<b>drop()</b>-funktio pudottaa halutut muuttujat pois datasta.

```python
# Tietyt rivit suodataan näin 
df[4:6]
```

```python
# ne joilla palkka on suurempi kuin 4000
df[df['palkka']>4000]
```

```python
# naiset joiden palkkatyytyväisyys on alle 3
#  = kun sijoitetaan jotakin, == kun verrataan jotakin , & = ja symbooli
df[(df['sukup']==2)&(df['palkkat']<3)]
```

```python
# katsotaan ne jotka ovat käyttäneet työterveyshuoltoa ja lomaosakkeita
# notnull funktiolla voi katsoa ei tyhjien määrät
# isnull funktiolla voi katsoa missä on tyhjiä arvoja
df[(df['työterv'].notnull())&(df['lomaosa'].notnull())]
```

```python
# Näin näytetään vain tietyt sarakkeet (eli suodatetaan sarakkeita pois) 
# Haluan listan tyytyväisyys ja käyttösarakkeista vaan ne joilla käyttö >= 3
df[['tyytyväisyys','käyttö']][df['käyttö']>=3]
```

```python
# Vastaavasti suodatetaan palkka ja palkkatyytyväisyys palkan mukaan sortattuna suuruusjärjestykseen ylhäältä alas
# sort_values( by kertoo minkämukaan sortataan ) 
# Tämä on hyvä esim. miten pythonissa kaikki on olioita. 
# Yleensä kaikki temput mitä tehdään palauttaa jonkin olion ja sille oliolle voidaan tehdä lisäjuttuja
# Eli ensin otetaan dataframesta palkka ja palkkat sarakkeet ja siitä tulle yksi olio 
# ja pisteen jälkeen kerrotaan mitä sille oliolle tehdään = sortataan arvot palkat mukaan ja siitä
# tulee oma olionsa ja sitten pisteen jälkeen kerrotaan mitä tälle oliolle tehdään 
# ja tässä esimerkissä kerrotaan että näytetään 5 ensimmäistä riviä.
# Eli näin voi ketjuttaa komentoja peräkkäin
# lisäparametrillä ascending=False - saadaan suurimmat ensin 
# Huom ! Tämä mitä nyt näkyy ei jää mihinkään talteen -- inplace=True on se lisäparametri joka tallentaisi tiedon.
df[['palkka','palkkat']].sort_values(by='palkka', ascending=False).head()
```

```python
# Jos muutama sarake vain pudotetaan pois, on sopiva komento drop jolla kerrotaan ne sarakket 
# jotka pudotetaan pois.   Huom ! Tässä pitää muistaa kertoa axis=1 eli mistä suunasta pudotetaan.   
df.drop(['nro','ikä','palveluv','palkka'], axis=1).head()
```
